```python
from sklearn import datasets
from scipy import stats as kde

import numpy as np
import matplotlib.pyplot as plt
iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names
dataX = []
dataY = []
data = []

# divide the data in the three calsses
for x in X:
    dataX.append(x[2])
    dataY.append(x[3])
    data.append([x[2],x[3]])
    
label1_x =[]  
label1_y =[] 
label1 = []
for i in range(50):
    label1_x.append(data[i][0])
    label1_y.append(data[i][1])
    label1.append(data[i])
label2_x =[] 
label2_y =[] 
label2 = []
for i in range(50,100):
    label2_x.append(data[i][0])
    label2_y.append(data[i][1])
    label2.append(data[i])
label3_x =[]
label3_y =[]
label3 = []
for i in range(100,150):
    label3_x.append(data[i][0])
    label3_y.append(data[i][1])
    label3.append(data[i])
    
    
    
labelX = []
labelY = []    
for x in X:
    labelX.append(x[2])
    labelY.append(x[3])


data = np.array(data) 
label1 = np.array(label1)
label2 = np.array(label2)
label3 = np.array(label3)


plt.plot(label1_x,label1_y,'*r')
plt.plot(label2_x,label2_y,'*b')
plt.plot(label3_x,label3_y,'*g')
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib.mlab import bivariate_normal
from mpl_toolkits.mplot3d import Axes3D

## EStimation for the first class ##
xmin = min(label1_x)
xmax = max(label1_x)
ymin = min(label1_y)
ymax = max(label2_y)
X, Y = np.mgrid[-1:3:50j, -1:2:50j]



bw_method = [0.1,1,3,6,8]
lim = [12,5,1,0.4,0.2]
for bw, lim  in zip(bw_method,lim):
    scalar = kde.gaussian_kde(label1.T, bw)
    fig = plt.figure(figsize=(10, 7))
    ax = fig.gca(projection='3d')
    Z = scalar(np.array([X.ravel(),Y.ravel()]))
    Z = Z.reshape(50,50)
    surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm,
            linewidth=0, antialiased=False)

    ax.set_zlim(0,lim)
    ax.zaxis.set_major_locator(plt.LinearLocator(10))
    ax.zaxis.set_major_formatter(plt.FormatStrFormatter('%.02f'))
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('p(x)')

    plt.title('Gaussian kernel, ancho de ventana = ' + str(bw))
    fig.colorbar(surf, shrink=0.5, aspect=7, cmap=plt.cm.coolwarm)
    plt.savefig('kernel_' + str(bw) +'.jpg')
    plt.show()
```

```python
## EStimation for the second class ##

X, Y = np.mgrid[2:6:50j, 0.5:2.3:50j]



scalar = kde.gaussian_kde(label2.T, bw_method='scott')
fig = plt.figure(figsize=(10, 7))
ax = fig.gca(projection='3d')
Z = scalar(np.array([X.ravel(),Y.ravel()]))
Z = Z.reshape(50,50)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm,
        linewidth=0, antialiased=False)

ax.set_zlim(0,3)
ax.zaxis.set_major_locator(plt.LinearLocator(10))
ax.zaxis.set_major_formatter(plt.FormatStrFormatter('%.02f'))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('p(x)')

plt.title('Gaussian kernel, bw_method ')
fig.colorbar(surf, shrink=0.5, aspect=7, cmap=plt.cm.coolwarm)
plt.savefig('kernel_scott.jpg')
plt.show()
```

```python
## EStimation for the third class ##

X, Y = np.mgrid[2:7:50j, 0:3:50j]



scalar = kde.gaussian_kde(label2.T, bw_method='silverman')
fig = plt.figure(figsize=(10, 7))
ax = fig.gca(projection='3d')
Z = scalar(np.array([X.ravel(),Y.ravel()]))
Z = Z.reshape(50,50)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm,
        linewidth=0, antialiased=False)

ax.set_zlim(0,3)
ax.zaxis.set_major_locator(plt.LinearLocator(10))
ax.zaxis.set_major_formatter(plt.FormatStrFormatter('%.02f'))
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('p(x)')

plt.title('Gaussian kernel, bw_method ')
fig.colorbar(surf, shrink=0.5, aspect=7, cmap=plt.cm.coolwarm)
plt.savefig('kernel_silverman.jpg')
plt.show()
```
